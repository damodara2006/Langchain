**PROMT_TEMPLATE**

```python
from langchain_core.prompts import PromptTemplate

user = input("Enter topic")
data = input("Enter promt")

dynamic_promt = PromptTemplate.from_template("{data} {topic}")
promt = dynamic_promt.invoke({"topic":user, "data":data})
print(promt)
model.invoke(promt).content
```

**CHAT_PROMT_TEMPLATE**

```python
from langchain_core.prompts import ChatPromptTemplate

chat_data = input("Enter you promt")

dynamic_promt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a doctor"),
        ("user","{data}")
    ]
)
for chunk in model.stream(dynamic_promt.invoke({"data":chat_data})):
    print(chunk.content)
```
