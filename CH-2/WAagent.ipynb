```python
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent
from langchain.chat_models import init_chat_model
import os

os.getenv("GOOGLE_API_KEY")
```

```python
ai_model = init_chat_model("google_genai:gemini-2.5-flash-lite", temperature=1)
```

```python
from langchain.tools import tool
from langchain.agents import create_agent  # or your correct agent creator

@tool("calculator", description="Performs arithmetic calculations. Use this for any math problems.")
def add_numbers(expression: str) -> str:
    """Evaluate mathematical expressions."""
    return str(eval(expression))

tools = [add_numbers]

agent = create_agent(
    model=ai_model,
    tools=tools
)

agent.invoke({"expression":"10 + 11"})
```
